### Rotina para extrair dados da Pnad 2015

#### Variáveis de pessoas


```python
# Primeiramente colocamos o arquivos com os dados, nesse caso são os dados de pessoas
# Em colspecs colocamos as posíções das variáveis desejadas, essas posições encontramos no dicionário da Pnad
# *No dicionário a variável ano está na posição 1-4, aqui colocamos (0,4) porque o jupyter começa a contar do zero
# e o intervalo é aberto no último valor, ou seja, quando escolhemos (0,4) teremos os valores '0,1,2,3' e não o '4'
# No vetor names escolhemos os nomes da variáveis de acordo com a posição em colspecs

pnad2015 = pd.read_fwf('PES2015.txt', 
                       colspecs=[(0,4), (4,6), (4,12), (17,18), (26,29), (32,33), (44,45), (153,154), 
                                 (109,113), (306,307), (309,310), (370,372), (374,376), (376,378), 
                                 (415,416), (679,681), (683,685), (699,701), (701,713), (765,766), 
                                 (766,767), (778,779), (790,795)],
names=['ano','uf', 'controle', 'sexo', 'idade', 'cor', 'nasmun', 'trabnasema', 'trabprinc', 'ocupnagr', 
       'puboupriv', 'htrabporsem', 'numdeanostrab', 'numdemestrab', 'sind', 'anosest', 'cartass', 
       'grupoocup', 'rendm', 'met', 'rurb', 'grupanosest', 'pesop'],
       encoding='iso-8859-1')
```

```python
# Aqui transformei a var de rendimento em var numérica
pnad2015['rendm'] = pd.to_numeric(pnad2015['rendm'], errors ='coerce')
```
